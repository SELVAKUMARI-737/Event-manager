```python
df['FuelType'].fillna(df['FuelType'].mode()[0],inplace=True)
df.info()
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df['MetColor'].fillna(df['MetColor'].mode()[0],inplace=True)
df.info()
```

```python
df.to_csv("cleanedToyato.csv",index=False)
```
